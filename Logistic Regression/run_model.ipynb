```python
from PIL import Image as img
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import random
import math

# Local modules
from extract_data import *
from model import *
```

```python
theta = np.zeros((n,1))+0.000001
th = theta
# print(theta)
alpha = 0.0001
x = train_x
y = train_y
print(x.shape,y.shape,theta.shape,'alpha =',alpha)
print(J(g(theta.transpose() @ x),y))
theta = gd(theta,x,y,alpha,200,graph=1,cost=1)
print(J(g(theta.transpose() @ x),y))
# print(g(theta.transpose() @ x))
```

```python
n = n+1 #"""for bias term 1 added to n"""
m = m_train
theta = np.zeros((n,1))+0.000001
# x = np.random.randint(l,r,(n,m))/r
# y = np.random.randint(0,2,(1,m))
y = train_y.copy()
xx = (train_x/255).copy()
print(y.shape,xx.shape)
x = np.vstack((xx,np.ones((1,m))))
```

```python
"""Alpha VS Cost"""
# th = np.zeros((n,1))+0.000001
alpha = np.linspace(0.0001,1,10)
cost = []
for a in alpha:
    theta = np.zeros((n,1))+0.000001
    theta = gd(theta,x,y,a,20)
    # print(theta,end=',')
    cost += [J(g(theta.transpose() @ x),y)]
# print(cost)
plt.close()
plt.plot(alpha,cost,'r.-')
plt.show()
```

```python
def predict(x,y,theta):
    predict_y = g(theta.T @ x)
    predict_y[predict_y>=0.5]=1
    predict_y[predict_y<0.5]=0
    # print(predict_y==y)
    return len(predict_y[predict_y==y])*100/m
```
